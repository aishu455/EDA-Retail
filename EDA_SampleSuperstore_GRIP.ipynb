```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True,bw_method='scott')
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True,bw_method='scott')
plt.xlim([-100,1000])
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.char.array(['New york city','Los Angeles','Philadelphia','San Francisco','Seattle'])
y = np.array([915,747,537,510,428])
colors = ['yellowgreen','red','gold','lightskyblue','white']
porcent = 100.*y/y.sum()

patches, texts = plt.pie(y, colors=colors, startangle=90, radius=2.2)
labels = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(x, porcent)]

sort_legend = True
if sort_legend:
    patches, labels, dummy =  zip(*sorted(zip(patches, labels, y),
                                          key=lambda x: x[2],
                                          reverse=True))

plt.legend(patches, labels, loc='center right', bbox_to_anchor=(-0.1, 1.),
           fontsize=16)

plt.savefig('piechart.png', bbox_inches='tight')
```
